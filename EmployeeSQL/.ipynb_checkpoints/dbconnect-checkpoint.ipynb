```python
import os
import pandas as pd

from dotenv import load_dotenv
load_dotenv()

#import sqlalchemy
from sqlalchemy import create_engine
```

```python
# Build connect string with credentials hidden
# Connect to db engine
uname = os.environ.get('db_username')
pword = os.environ.get('db_password')
connect_string = f"postgresql://{uname}:{pword}@localhost:5432/movierentals"
engine = create_engine(connect_string)
connection = engine.connect()
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_sql.html
actor_df = pd.read_sql("select * from \"actor\"", connection);

actor_df
```

```python
actor_df.info()
```

```python
## https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_sql_table.html#pandas.read_sql_table

table_df = pd.read_sql_table('actor', con=connection)
table_df
```
